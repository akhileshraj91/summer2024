```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
energy_file = pd.read_csv('energy.csv')
power_file = pd.read_csv('measured_power.csv')
```

```python
power = {}
power['geopm_0'] = power_file[power_file.scope == 0]
power['geopm_1'] = power_file[power_file.scope == 1]
energy = {}
energy['geopm_0'] = energy_file[energy_file.scope == 0]
energy['geopm_1'] = energy_file[energy_file.scope == 1]
```

```python
# print(power)
# print(energy)
power_calc = {}
for sensor in energy.keys():
    # print(sensor)
    # print(energy[sensor])
    power_calc[sensor]= energy[sensor].value.diff()/energy[sensor].time.diff()
# print(power_calc.keys())
power_calc['elapsed_time'] = energy['geopm_0'].time - energy['geopm_1'].time.iloc[0]
power['elapsed_time'] = power['geopm_0'].time - power['geopm_1'].time.iloc[0]
# power_df = pd.DataFrame(power)
# power_calc_df = pd.DataFrame(power_calc)

power_calc['elapsed_time'] = pd.to_numeric(power_calc['elapsed_time'], errors='coerce')
power['elapsed_time'] = pd.to_numeric(power['elapsed_time'], errors='coerce')
```

```python
fig,axs = plt.subplots(1,1)
lw = 2
axs.scatter(power['elapsed_time'],power['geopm_0'].value,label='power_sensor_0')
axs.scatter(power['elapsed_time'],power['geopm_1'].value, label='power_sensor_1')
axs.scatter(power_calc['elapsed_time'],power_calc['geopm_0'], label = 'calc_power_0')
axs.scatter(power_calc['elapsed_time'],power_calc['geopm_1'], label = 'calc_power_1')
axs.grid(True)
axs.legend()
plt.tight_layout()
plt.savefig('output.pdf')
```
